```python

X = np.linspace(0, 100, 100)

X2 = X**2

fig, ax1 = plt.subplots(figsize=(8, 4))
ax1.plot(X, X2, label='X^2', color='b')
ax1.set_xlabel('X')
ax1.set_ylabel('X^2', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.grid(True)
ax1.legend(loc='upper left')

exp_X = np.exp(X)

ax2 = ax1.twinx() 
ax2.scatter(X, exp_X, label='exp(X)', color='r', marker='x')
ax2.set_ylabel('exp(X)', color='r')
ax2.tick_params(axis='y', labelcolor='r')
ax2.legend(loc='upper right')

plt.title('Plot of X vs. X^2 and X vs. exp(X)')
plt.show()
```

Both plots worked for me but I would assume price wouldn't work because of the $ and commas

```python

car_sales['Price'] =car_sales['Price'].str.replace(r'[\$\,]', '', regex=True)

car_sales['Price'] = car_sales['Price'].astype('int64')
```

```python
car_sales.head()
```

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales
```

```python
import pandas as pd
import datetime

today = datetime.date.today()
date_range = pd.date_range(start=today, periods=len(car_sales))

car_sales['Sale Date'] = date_range
```

```python
car_sales.head()
```

```python

plt.figure(figsize=(10, 6)) 
plt.plot(car_sales['Total Sales'], car_sales['Price'], label='Price')


plt.xlabel('Price')
plt.ylabel('Total Sale')
plt.title('Price vs total Sales Line Plot')

plt.legend()
plt.show()
```

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Pricee Scatter Plot')
plt.show()
```

```python
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer Bar Plot')
```

```python
heat_map=car_sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
cols=['Total Sales','Price','Odometer (KM)']
sns.heatmap(heat_map, xticklabels=cols, yticklabels=cols, annot=True)
```

This heatmap shows that there is a negative correlation between the three because they are less than one and some involve a negative number
